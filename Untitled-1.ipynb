```python
import pandas as pd
from pathlib import Path

current_file_path = Path().resolve() / "Untitled-1.ipynb"
df = pd.read_csv('wine_quality_merged.csv')
df = df.sample(n=5150, random_state=41).reset_index(drop=True)

print(df.head())
print(len(df))

print(df.info())



df['type'] = df['type'].map({'red': 1, 'white': 0}) #Normalizacion datos - conversion de categorica a numerica (CC)
print(df.head())

X = df.drop('quality', axis=1)
y = df['quality']

from sklearn.preprocessing import StandardScaler, LabelEncoder, SMOTE
from sklearn.ensemble import IsolationForest

import numpy as np

def prepare_data(df, scenario='base', random_state=42):
    """
    Prepara datos según el escenario especificado
    
    Escenarios:
    1. base: Sin modificaciones
    2. scaled: Con escalado
    3. no_outliers: Sin outliers
    4. scaled_no_outliers: Escalado + sin outliers
    5. balanced: Con balanceo de clases
    6. scaled_balanced: Escalado + balanceado
    7. optimized: Escalado + sin outliers + balanceado
    8. raw: Sin limpieza adicional
    """
    
    df_processed = df.copy()
    
    # Separar features y target
    if 'quality' in df_processed.columns:
        X = df_processed.drop(['quality', 'type'], axis=1, errors='ignore')
        y = (df_processed['quality'] >= 6).astype(int)  # Clasificación binaria
    else:
        raise ValueError("No se encontró columna 'quality'")
    
    # Aplicar transformaciones según el escenario
    
    if scenario in ['base', 'raw']:
        # Sin modificaciones
        pass
    
    elif scenario == 'no_outliers':
        # Eliminar outliers usando IsolationForest
        iso_forest = IsolationForest(contamination=0.05, random_state=random_state)
        outlier_mask = iso_forest.fit_predict(X) == 1
        X = X[outlier_mask]
        y = y[outlier_mask]
    
    elif scenario == 'scaled':
        # Escalado
        scaler = StandardScaler()
        X = pd.DataFrame(
            scaler.fit_transform(X),
            columns=X.columns
        )
    
    elif scenario == 'scaled_no_outliers':
        # Outliers primero, luego escalar
        iso_forest = IsolationForest(contamination=0.05, random_state=random_state)
        outlier_mask = iso_forest.fit_predict(X) == 1
        X = X[outlier_mask]
        y = y[outlier_mask]
        
        scaler = StandardScaler()
        X = pd.DataFrame(
            scaler.fit_transform(X),
            columns=X.columns
        )
    
    elif scenario == 'balanced':
        # Balanceo SMOTE
        smote = SMOTE(random_state=random_state)
        X, y = smote.fit_resample(X, y)
        X = pd.DataFrame(X, columns=X.columns)
    
    elif scenario == 'scaled_balanced':
        # Escalar primero, luego balancear
        scaler = StandardScaler()
        X = pd.DataFrame(
            scaler.fit_transform(X),
            columns=X.columns
        )
        smote = SMOTE(random_state=random_state)
        X, y = smote.fit_resample(X, y)
        X = pd.DataFrame(X, columns=X.columns)
    
    elif scenario == 'optimized':
        # Pipeline completo: outliers → escalar → balancear
        iso_forest = IsolationForest(contamination=0.05, random_state=random_state)
        outlier_mask = iso_forest.fit_predict(X) == 1
        X = X[outlier_mask]
        y = y[outlier_mask]
        
        scaler = StandardScaler()
        X = pd.DataFrame(
            scaler.fit_transform(X),
            columns=X.columns
        )
        
        smote = SMOTE(random_state=random_state)
        X, y = smote.fit_resample(X, y)
        X = pd.DataFrame(X, columns=X.columns)
    
    return X, y


scenario = 'base'  # Cambia el escenario según lo que quieras probar
X, y = prepare_data(df, scenario=scenario)
print(f"\n✅ Datos preparados: {X.shape[0]} muestras, {X.shape[1]} features")
print(f"   Distribución: {np.bincount(y).tolist()}")
```
